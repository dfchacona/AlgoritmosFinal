# Problema de cubrimiento de un tablero 3 xn con chas de domino


```python
# Tomado de https://blog.csdn.net/y12345678904/article/details/52150817

a = [0]*120
```

```python
def func(n):
    i= 0
    sum = 0
    if n == 0:
        return 1
    if a[n] != 0:
        return a[n]
    i = n - 2
    while (i>=0):
        if i == n-2:
            sum +=3 * func(i)
        else:
            sum += 2 * func(i)
        i-= 2
    a[n] = sum
    return sum
```

```python
if __name__ == "__main__":
    n = 10
    print(func(n))
```
